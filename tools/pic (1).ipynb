```python
import pylab
import numpy as np
```

```python
path = '/home/jovyan/res/pic_guid.jpg'
im = pylab.imread(path)
im_1 = im[0:340,2350:2625]

im_n = np.zeros(im.shape,dtype=im.dtype)
print(im_n.shape)
print(im[0][0]) #(52,54,67)
#im[0:340,2350:2625] = im[0:340,1750:2225]
i = 0
for y in im:
    j = 0
    if i >= 340:
        for x in y:
            im_n[i,j,0]=im[i,j,0]
            im_n[i,j,1]=im[i,j,1]
            im_n[i,j,2]=im[i,j,2]
            j = j+1
    else:
        for x in y:
            if j > 2350:
                im_n[i,j,0] = 52
                im_n[i,j,1] = 54
                im_n[i,j,2] = 67
                j = j+1
            else:
                im_n[i,j,0] = im[i,j,0]
                im_n[i,j,1] = im[i,j,1]
                im_n[i,j,2] = im[i,j,2]
                j = j+1
    i = i+1

pylab.imshow(im_n)

pylab.show()
```

```python
pylab.imsave('pic.png',im_n)
```
